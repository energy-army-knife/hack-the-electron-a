### Evolution of power by month

```python
months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
mean_power_months = np.zeros((24, df_power.shape[1]-2))

n = 0
for year in ["2016", "2017", "2018"]:
    for month in months:
        date_upper = "{0}-{1}-01".format(year, month)
        date_lower = "{0}-{1}-31".format(year, month)
        filered_data = df_power[(df_power['Time'] > date_upper) & (df_power['Time'] < date_lower)]
        if filered_data.shape[0] > 1:
            filered_data.drop(filered_data.columns[[0, 1]], axis=1, inplace=True)
            mean_power_months[n,:] = filered_data.mean().to_numpy()
            n += 1      
```

```python
cmap = plt.get_cmap('gnuplot')
colors = [cmap(i) for i in np.linspace(0, 1, mean_power_months.shape[1])]

figure(num=None, figsize=(20, 10), facecolor='w', edgecolor='k')
for i, color in enumerate(colors, start=1):
    plt.plot(range(24), mean_power_months[:,i-1], color=color)
plt.legend(loc='best')
plt.show()
```
