### Evolution of power by month


```python
database_group_by_month = df_dataset.groupby(df_dataset.index.strftime('%Y-%B')).mean()
```

```python

database_group_by_month.shape
```

```python
figure(num=None, figsize=(20, 10))
x = plt.plot(database_group_by_month)
```

```python
meter_info = MetersInformation("../resources/dataset_index.csv")
figure(num=None, figsize=(20, 10), facecolor='w', edgecolor='k')
colors = ["red", "blue", "black"]
for i, tariff in enumerate([TariffType.SIMPLE, TariffType.THREE_PERIOD, TariffType.TWO_PERIOD]):
    meters_ids = meter_info.get_meters_id_by_tariff(tariff)
    plt.plot(database_group_by_month[set(meters_ids) & set(database_group_by_month.columns)], color=colors[i])
```

```python
building_data_loader = BuildingDataLoader("../resources/HackTheElectron dataset support data/"
                                              "Building Info-Table 1.csv")

figure(num=None, figsize=(20, 10), facecolor='w', edgecolor='k')
building_types = ["Small", "Medium", "Large"]
for i, building_type in enumerate(building_types):
    meters_ids = building_data_loader.get_meters_id_by_building_type(building_type)
    plt.plot(database_group_by_month[set(meters_ids) & set(database_group_by_month.columns)], color=colors[i])
```
